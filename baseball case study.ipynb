# Separating the data

```python
#independent column
x=df.drop(columns='W',axis=1)
```

```python
x
```

```python
x.shape
```

```python
y = df['W']
```

```python
y
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression,Lasso
from sklearn.metrics import mean_absolute_error, accuracy_score,mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
#splitting the data
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=31,test_size=0.2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# 1. Linear_regression model

```python
def df (model,x_train,y_train):
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    cv_score=cross_val_score(model,x_train,y_train, scoring='neg_mean_squared_error',cv=5)
    cv_score=np.abs(np.mean(cv_score))
    print('Model Report')
    print('MSE:',mean_squared_error(y_test,pred))
    print('CV score:',cv_score)
    print("R2 score : %.2f" % r2_score(y_test,pred))
    print('MAE:', mean_absolute_error(y_test,pred))
    print('RMSE:', np.sqrt(mean_squared_error(y_test,pred)))
```

```python
model1=LinearRegression(normalize=True)
df(model1,x_train,y_train)
coef=pd.Series(model1.coef_,x.columns).sort_values()
coef.plot(kind='bar',title='model coefficient')
```

Recall that the average number of wins is about 80 wins. On average, the model is off by about 2.73 wins.

# 2. Ridge regression model

```python
from sklearn.linear_model import RidgeCV
```

```python
model2= RidgeCV(alphas=(0.01,0.1,1,10),normalize=True)
df(model2,x_train,y_train)
coef=pd.Series(model2.coef_,x.columns).sort_values()
coef.plot(kind='bar',title='model coefficient')
```

Ridge regression model is slightly better than linear regression model because it has less mean absolute error.

# 3. Decision tree regressor model

```python
from sklearn.tree import DecisionTreeRegressor
model3=DecisionTreeRegressor()
df(model3,x_train,y_train)
coef=pd.Series(model3.feature_importances_,x.columns).sort_values(ascending=False)
coef.plot(kind='bar',title='Feature Importance') 
```

# 4. Lasso model

```python
model4=Lasso(normalize=True)
df(model4,x_train,y_train)
coef=pd.Series(model4.coef_,x.columns).sort_values()
coef.plot(kind='bar',title='model coefficient')
```

# 5. Random forest model

```python
y.value_counts()
```

```python
from sklearn.ensemble import RandomForestRegressor
model5=RandomForestRegressor()
df(model5,x_train,y_train)
coef=pd.Series(model5.feature_importances_,x.columns).sort_values(ascending=False)
coef.plot(kind='bar',title='Feature Importance')
```

# 6.ExtraTree regressor model

```python
from sklearn.ensemble import ExtraTreesRegressor
model3=ExtraTreesRegressor()
df(model3,x_train,y_train)
coef=pd.Series(model3.feature_importances_,x.columns).sort_values(ascending=False)
coef.plot(kind='bar',title='Feature Importance') 
```

From the above models random forest model is doing best job as errors are low in it.
